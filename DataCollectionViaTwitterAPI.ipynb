```python
def get_tweets(api, query, max_results, max_requests):
    tweets = []

    for _ in range(max_requests):
        # Perform a search request
        response = api.search_recent_tweets(query,
                                            max_results=max_results,
                                            tweet_fields=['id', 'text', 'created_at', 'lang', 'geo', 'public_metrics', 'context_annotations', 'entities'],
                                            user_fields=['id', 'username', 'description', 'created_at', 'entities', 'location', 'verified', 'public_metrics'],
                                            place_fields=['full_name', 'id', 'country', 'country_code', 'geo'],
                                            expansions=['author_id', 'geo.place_id','referenced_tweets.id','attachments.media_keys'],
                                            end_time='2023-11-21T10:42:27.000Z')  

        # Check if there are more tweets
        if 'data' in response:
            tweets += response['data']

        # Check if there is a next page
        if 'meta' in response and 'next_token' in response['meta']:
            next_token = response['meta']['next_token']
        else:
            break  

    return tweets, response.get('includes', {})

# Initialize Tweepy Client
client = tweepy.Client(bearer_token=bearer_token,
                       consumer_key=consumer_key,
                       consumer_secret=consumer_secret,
                       access_token=access_token,
                       access_token_secret=access_secret,
                       return_type=dict,
                       wait_on_rate_limit=True)

# Specify the query and get tweets
# query = 'SamAltman OR OpenAI OR "Satya Nadella" OR "Sam Altman" OR "Greg Brockman" OR #OpenAI OR #Microsoft OR "Adam D‚ÄôAngelo" OR #SamAltman'
query = 'SamAltman OR OpenAI -"RT"'
all_tweets, includes_data = get_tweets(client, query, max_results=50, max_requests=5)

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame.from_records(all_tweets)

# Access additional data from includes
user_data = includes_data.get('users', {})
place_data = includes_data.get('places', {}).get('places', [])

# Now, you can merge this additional data into your DataFrame using the user_id and place_id
if user_data:
    df = pd.merge(df, pd.DataFrame.from_records(user_data), left_on='author_id', right_on='id', how='left', suffixes=('', '_user'))
if place_data:
    df = pd.merge(df, pd.DataFrame.from_records(place_data), left_on='geo.place_id', right_on='id', how='left', suffixes=('', '_place'))

# Save the DataFrame to a CSV file
df.to_csv('tweets_OpenAI_6.csv', index=False)
```
