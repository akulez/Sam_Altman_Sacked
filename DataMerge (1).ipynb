```python
en_df = merged_df[merged_df['lang']=='en']
```

```python
import emoji
import regex
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Replace emojis with their descriptions
def replace_emojis_with_meaning(text):
    return regex.sub(r'\p{So}', lambda x: f' {emoji.demojize(x.group())} ', text)

# Perform data cleaning
def clean_text(text):
    # Convert to lowercase
    text = text.lower()

    # Replace emojis with their meanings
    text = replace_emojis_with_meaning(text)

    # Remove punctuation
    text = ''.join([char for char in text if char not in string.punctuation])

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text)
    text = ' '.join([word for word in words if word.lower() not in stop_words])

    # Lemmatization 
    lemmatizer = WordNetLemmatizer()
    text = ' '.join([lemmatizer.lemmatize(word) for word in word_tokenize(text)])

    return text

en_df['text'] = en_df['text'].apply(clean_text)
```

```python
from wordcloud import WordCloud

all_text = ' '.join(en_df['text'])

# Word Cloud for most used words
plt.figure(figsize=(12, 6))
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Most Used Words')
plt.show()
```

```python
from nltk.probability import FreqDist

word_tokens = word_tokenize(all_text)
fdist = FreqDist(word_tokens)
fdist.plot(30, cumulative=False, title='Top 30 Words by Frequency')
```

```python
from nltk import ngrams

n = 2  # 
n_grams = ngrams(word_tokens, n)
fdist_ngrams = FreqDist(n_grams)
fdist_ngrams.plot(30, cumulative=False, title=f'Top 30 {n}-grams by Frequency')
```

```python

en_df['doc_length'] = en_df['text'].apply(lambda x: len(word_tokenize(x)))
plt.hist(en_df['doc_length'], bins=30, edgecolor='k')
plt.title('Distribution of Document Lengths')
plt.xlabel('Number of Words')
plt.ylabel('Number of Documents')
plt.show()
```

```python
target_word = 'return'  
context_words = [word_tokens[i-1] + ' ' + word_tokens[i+1] for i, word in enumerate(word_tokens) if word == target_word and i > 0 and i < len(word_tokens)-1]
fdist_context = FreqDist(context_words)
fdist_context.plot(10, cumulative=False, title=f'Contextual Words for "{target_word}"')
```
